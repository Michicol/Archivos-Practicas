```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sy
```

```python
def Ps(beta,S):
    A = 2 * (sy.special.gamma((beta+2/2)))**(beta + 1) / (sy.special.gamma((beta + 1)/2))**(beta+2)
    B = (sy.special.gamma((beta+2)/2)/sy.special.gamma((beta+1)/2))**2

    return A* S**beta * np.exp(-B*S**2)
```

```python
def Pr(beta,r):
    C = (3**(3*(1+beta)/2)*(sy.special.gamma(1+beta/2)**2))/(2*np.pi*sy.special.gamma(1+beta))
    return C * (r + r**2)**beta / (1 + r + r**2)**(1+3*beta/2)
```

```python
def Plam(beta,x,n):
    c = 1/(beta*n*np.pi)
    return c* np.sqrt(2*beta*n - x**2)
```

# GOE

Para construir una matriz Hamiltoniana con entradas aleatorias de $N\times N$, donde los elementos de su diagonal tiene una distribución $N(0,1)$ es decir $\mu=0$ y $σ^2 = 1$, y por otra parte sus elementos distintos de la diagonal deben tener una distribución $N(0,1/2)$ es decir $\mu=0$ y $\sigma^2 = 1/2$, además dicha matriz debe ser simetrica.

```python
def MatHamiltoniana(n):
  #Se crea una matriz Y con entradas aleatorias con una distribucion norma
  # con mu = 0 y sigma^2 = 1
  Y = np.random.normal(0,1,(n,n))
  #Ahora se construye la matriz hamiltoniana, con una distribucion N(0,1) en
  #la diagonal y fuera de la diagonal N(0,1/2), y además simetrica
  H = 0.5 * (Y + Y.T)
  return H
```

```python
n = 10000
```

```python
H = MatHamiltoniana(n) #Se construye la matriz hamiltoniana de 1000x1000
```

```python
eigenvaloresGOE = np.linalg.eigvalsh(H)
eigenvaloresGOENorm = eigenvaloresGOE / np.sqrt(n)
```
