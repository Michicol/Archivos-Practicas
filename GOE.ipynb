```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matricesAleatorias import matham as mh
```

```python
def P_s(x):
    return (np.pi/2) * x* np.exp((-4*x**2)/4)
```

```python
n = 10000
H = mh.MatrizHamiltoniana(n)
```

```python
HGOE = H.GOE()
```

```python
eigvalsGOE = np.linalg.eigvalsh(HGOE)
eigvalsGOEnorm = eigvalsGOE /np.sqrt(n)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(eigvalsGOE)
ax[0].set_title('Eigenvalores sin normalizar')
ax[0].set_xlabel('N')
ax[0].set_ylabel('E')

ax[1].plot(eigvalsGOEnorm)
ax[1].set_title('Eigenvalores normalizados')
ax[1].set_xlabel('N')
ax[1].set_ylabel('E')
```

```python
densidad = mh.DensidadesHam(n,1)
```

```python
xlam = np.linspace(eigvalsGOE.min(),eigvalsGOE.max(),n)
xlamnor = np.linspace(eigvalsGOEnorm.min(),eigvalsGOEnorm.max(),n)
```

```python
plam = densidad.Plam(xlam)
plamnor = densidad.Plam(xlamnor)
```
