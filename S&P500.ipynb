```python
AMAT_IT = df['AMAT_IT']
```

```python
AMAT_ITnp = np.array(AMAT_IT)
```

```python
plt.figure()

plt.hist(AMAT_ITnp[:300], bins=20,density=True,lw=0.5,edgecolor='white')

plt.show()
```

```python
color = ['red','blue','orange','purple','green']
plt.figure()

for i in range(6):
    plt.hist(AMAT_ITnp[i*100:i*100+100], bins=20,histtype='step', density=True,lw=1,alpha=1)

plt.show()
```

```python
retornos = np.diff(AMAT_ITnp)/AMAT_ITnp[:-1]
```
