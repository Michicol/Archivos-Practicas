```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
df = pd.read_csv("Nikkei 225.csv",header=[0,1,2])#skiprows=1)
```

```python
df
```

```python
for i in range(3):
    # Assuming your DataFrame is already loaded as `df`
    current_columns = df.columns.get_level_values(i).tolist() # This is your list of column names
    new_column_names = {}  # Dictionary to store the changes
    current_category = None  # Variable to keep track of the current category

    # Loop through each column to determine its new name
    for col_name in current_columns:
        if 'Unnamed' not in col_name:
            # This is a category header
            current_category = col_name
            # The category header keeps its name, so no need to add to the rename dict
        else:
            # This is an 'Unnamed' column that should be renamed to the current category
            if current_category is not None:
                new_column_names[col_name] = current_category

    # Apply the renaming
    df.rename(columns=new_column_names, inplace=True)

# Display the result to verify
print(df.columns.tolist())
```

```python
Categoria = df.columns.get_level_values(0).tolist()
Industria = df.columns.get_level_values(1).tolist()
```

```python
Industria
```

```python
sectores = {}
```

```python
for i, j in zip(Categoria, Industria):
    sectores[j] = df.xs(j,level=1,axis=1).xs(i,level=0,axis=1).dropna()
    aux = sectores[j]
    aux['Code'] = aux['Code'].astype(int)
    aux['Code'] = aux['Code'].astype('str')+'.T'
```
