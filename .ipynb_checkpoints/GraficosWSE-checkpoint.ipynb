```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matricesAleatorias import matwishart as ws
from tqdm import tqdm
```

```python
Nmat = 1000
Tmat = 10000
mat = ws.MatrizWishart((Nmat,Tmat))
densidad = ws.Densidades((Nmat,Tmat),1000,4)
```

```python
WSEmat = mat.WSE()
```

```python
EvalsWSE = np.linalg.eigvalsh(WSEmat)
```

```python
EvalsWSE = EvalsWSE[::2]
```

```python
x, y = densidad.MarchenkoPastur()
```

```python
rWSE= (EvalsWSE[1:-1] - EvalsWSE[2:]) / (EvalsWSE[:-2] - EvalsWSE[1:-1])
```

```python
r = np.linspace(0,6,1000)
pr = densidad.Pr(r)
```

```python
nsamp = 100000
N = 100
T = 1000
```

```python
ensWSE = ws.MatrizWishart((N,T))
EnsEvalWSE = []
```

```python
for i in tqdm(range(nsamp)):
    WSE = ensWSE.WSE()
    eval = np.linalg.eigvalsh(WSE)
    EnsEvalWSE.append(eval)
```

```python
EnsEvalWSE = np.concatenate(EnsEvalWSE)
```

```python
EnsSinRep = []

for i in range(int(nsamp/N)):
    EvalSamp = EnsEvalWSE[0 + i*200:200 + i*200:2]
    EnsSinRep.append(EvalSamp)
```

```python
EnsSinRep = np.concatenate(EnsSinRep)
```

```python
rens = []
for i in range(int(nsamp/N)):
    EvalSamp = EnsSinRep[0+i*100:100 + i*100]
    rsamp = (EvalSamp[1:-1] - EvalSamp[2:]) / (EvalSamp[:-2] - EvalSamp[1:-1])
    rens.append(rsamp)
```

```python
rens = np.concatenate(rens)
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,5))

fig.suptitle(r'Eigenvalores de WSE $ \beta=4 $')

ax[0].plot(EvalsWSE,label='Eigenvalores completos')
ax[0].set_title('Eigenvalores normalizados matriz grande')
ax[0].grid(True,linestyle='--')
ax[0].legend()
ax[0].set_xlabel(r'$ n $')
ax[0].set_ylabel(r'$ \lambda $')

for i in range(int(nsamp/N)):
    ax[1].plot(EnsSinRep[0+i*100:100 + i*100])
ax[1].set_title('Eigenvalores normalizados del ensamble')
ax[1].grid(True,linestyle='--')
ax[1].set_xlabel(r'$ n $')
ax[1].set_ylabel(r'$ \lambda $')

plt.savefig('EigenvaloresWSE.png',dpi=300)
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,5))

fig.suptitle(r'Distribución de Marchenko-Pastur WSE $ \beta=4 $',fontsize=16)

ax[0].plot(x,y,lw=2)
ax[0].hist(EvalsWSE/(4*Nmat),bins=30,density=True,linewidth=0.5,edgecolor='white')
ax[0].grid(True,linestyle='--')
ax[0].set_title('Caso de una matriz muy grande')
ax[0].set_xlabel(r'$ \lambda $')
ax[0].set_ylabel(r'$ p(\lambda) $')

ax[1].plot(x,y,lw=2)
ax[1].hist(EnsSinRep/(4*N),bins=30,density=True,linewidth=0.5,edgecolor='white')
ax[1].grid(True,linestyle='--')
ax[1].set_title('Caso del Ensamble')
ax[1].set_xlabel(r'$ \lambda $')
ax[1].set_ylabel(r'$ p(\lambda) $')

plt.savefig('Marchenko-PasturWSE.png',dpi=300)
```

```python
fig, ax = plt.subplots(1,2,figsize=(17,5))

fig.suptitle(r'Distribución de razon de espaciamientos WSE $ \beta=4 $',fontsize=16)

ax[0].plot(r,pr,lw=2)
ax[0].hist(rWSE,bins=80,density=True,linewidth=0.5,edgecolor='white')
ax[0].grid(True,linestyle='--')
ax[0].set_title('Caso de una matriz muy grande')
ax[0].set_xlabel(r'$ r $')
ax[0].set_ylabel(r'$ p(r) $')
ax[0].set_xlim(0,10)

ax[1].plot(r,pr,lw=2)
ax[1].hist(rens,bins=80,density=True,linewidth=0.5,edgecolor='white')
ax[1].grid(True,linestyle='--')
ax[1].set_title('Caso del Ensamble')
ax[1].set_xlabel(r'$ r $')
ax[1].set_ylabel(r'$ p(r) $')
ax[1].set_xlim(0,10)

plt.savefig('prWSE.png',dpi=300)
```
