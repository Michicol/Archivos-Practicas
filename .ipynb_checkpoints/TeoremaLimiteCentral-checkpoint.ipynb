```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.uniform(0.0,1.0,(100,100000))
```

```python
x[0,:].sum()
```

```python
plt.figure()

plt.hist(x[0,:])

plt.show
```

```python
sns.histplot(data=x[0,:],kde=True)
```

```python
Normal = np.sum(x,axis=0) / x.shape[0]
```

```python
x.shape
```

```python
sns.histplot(data=Normal,kde=True)
```

```python
VarAleatorias = np.random.poisson(lam=np.linspace(0.6,1,100000),size=(100,100000))
```

```python
VarAleatorias.shape
```

```python
sns.histplot(VarAleatorias[0,:],kde=True)
```

```python
NormalPoisson = np.sum(VarAleatorias,axis=0) / VarAleatorias.shape[1]
```

```python
NormalPoisson.shape
```

```python
sns.histplot(data=NormalPoisson, kde=True)
```

```python
NormalPoisson.sum()
```

```python
VariableAleatoriaExponencial = np.random.exponential(1,(100,10000))
```

```python
sns.histplot(VariableAleatoriaExponencial[0,:],kde=True)
```

```python
NormalExponencial = np.sum(VariableAleatoriaExponencial,axis=0) / VariableAleatoriaExponencial.shape[1]
```

```python
sns.histplot(NormalExponencial,kde=True)
```
