```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos (igual que tu código)
np.random.seed(42)  # Opcional: para reproducibilidad
muestras = np.random.poisson(lam=3, size=(50, 100000))
medias_muestrales = np.mean(muestras, axis=0)

# Parámetros teóricos del TLC
media_teorica = 3
error_estandar = np.sqrt(3) / np.sqrt(50)  # σ/√n = √λ/√n

# Gráfico
plt.figure(figsize=(10, 6))
ax = sns.histplot(data=medias_muestrales, kde=True, stat='density', label='Medias muestrales')

# Superponer curva normal TEÓRICA
x = np.linspace(medias_muestrales.min(), medias_muestrales.max(), 200)
y = norm.pdf(x, loc=media_teorica, scale=error_estandar)
ax.plot(x, y, 'r--', linewidth=2, label='Normal teórica (TLC)')

plt.title('Teorema del Límite Central con Poisson(λ=3)\nMedia muestral (n=50) vs. Distribución Normal Teórica')
plt.xlabel('Media muestral')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde

# Generar datos exponenciales
np.random.seed(42)
exponential_data = np.random.exponential(1, (1000, 10000))

# Primer conjunto: primera fila (distribución exponencial)
data1 = exponential_data[0, :]

# Segundo conjunto: medias (distribución normal de medias)
data2 = np.mean(exponential_data, axis=0)

# Función para guardar datos con encabezados
def save_hist_kde(data, hist_file, kde_file):
    # Calcular histograma
    hist, bins = np.histogram(data, bins=50, density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2
    pd.DataFrame({'x': bin_centers, 'y': hist}).to_csv(hist_file, index=False)
    
    # Calcular KDE
    kde = gaussian_kde(data)
    x_kde = np.linspace(data.min(), data.max(), 500)
    y_kde = kde(x_kde)
    pd.DataFrame({'x': x_kde, 'y': y_kde}).to_csv(kde_file, index=False)

# Guardar datos con encabezados
save_hist_kde(data1, 'exponential_hist.dat', 'exponential_kde.dat')
save_hist_kde(data2, 'normal_hist.dat', 'normal_kde.dat')

# Imprimir rutas para verificación
print("Archivos guardados en:")
print(f"- exponential_hist.dat")
print(f"- exponential_kde.dat")
print(f"- normal_hist.dat")
print(f"- normal_kde.dat")
```
